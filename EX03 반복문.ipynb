# 반복문


```python
num = 0
while (num <= 3):
    if(num == 0):
        pass
    else:
        print(num)
    num += 1
```

```python
import random
result = random.randint(1, 50)
answer = 0

while(answer != result):
    answer = int(input("숫자를 입력하세요 >> "))

    if(answer > result):
        print(f'{answer}보다 작은 수 입니다.')
    elif(answer < result):
        print(f'{answer}보다 큰 수 입니다.')
    else:
        print("정답을 맞추셨습니다!")
```

```python
점심메뉴 = ["김치찌개", "순두부찌개", "짬뽕"]
for menu in range(len(점심메뉴)):
    print(점심메뉴[menu])
```

```python
import random

#1. randint 함수를 통해서 index number를 무작위로 가져온다.
result_menu = random.randint(0, len(점심메뉴)-1)
print(점심메뉴[result_menu])
```
