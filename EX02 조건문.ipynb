# Tuple
- 파이썬의 자료구조 형태 중 하나
- 순서가 있는 집합
- 소괄호()로 작성되어지며 내부의 데이터는 comma(,)로 구분됨
- 추가, 수정, 삭제 불가
- 튜플명 = (요소1, 요소2, 요소3, ...)
- 메모리 용량을 아낄 수 있음

````python
tuple1 = (0, 1, 2, 3, ('a', 'b', 'c'), 4)
print(tuple1[2])
print(tuple1[1:4])
````

## 실습1

````python
lst = ['딸기', '바나나', '수박', '체리', '포도']

word = input("검색할 문자를 입력하세요 >> ")
if(word in lst):
    print(f'{word}는 리스트에 {lst.index(word)}번째 인덱스에 있습니다.')
else:
    print(f'{word}는 리스트에 들어있지 않습니다.')
````

## 실습2

````python
string = "Hi, my name is yeongmin"

words = input("검색할 문자열을 입력하세요 : ")
if(words in string):
    print(f'{words}는 {string.count(words)}번 들어 있습니다.')
else:
    print(f'{words}는 문자열에 들어있지 않습니다.')
````
