```python
#COUNT 함수 활용
a = "안녕하세요용요요"
print(a.count("요"))
```

```python
#find 함수 활용 -> 찾을 없는 문자 있으면 -1 반환
b = "Bello"
print(b.find("H"))
```

```python
#index 함수 활용 -> 찾을 없는 문자 있으면 오류 발생
b = "Bello"
print(b.index("B"))
```

```python
# join() -> 문자열 삽입
name = "장영민"
print("❤️".join(name))
```

```python
# Upper(대문자), Lower(소문자)

c = "bye"
print(c.upper())

d = "ByE"
print(d.lower())
```

```python
#Lstrip, Rstrip, strip()
e = "   CGI   "
print(e.lstrip()) #왼쪽 공백 제거
print(e.rstrip()) #오른쪽 공백 제거
print(e.strip()) #양쪽 공백 제거
```

```python
# replace() -> 원하는 문자로 변경
b = "Hello"
print(b.replace('H', 'BB'))
```

```python
# split() -> 문자열 나누기
g = "첫번째문장, 두번째문장, 세번째문장"
print(g.split(","))
print(g.split(" "))
```

```python
num1 = 10
num2 = 7

# 나누기
print(num1 / num2)
# 몫
print(num1 // num2)
# 나머지
print(num1 % num2)
```

```python
# 문자열 더하기
str1 = "안녕"
str2 = "하세요"
str3 = "10"
num3 = 7

print(str1 + str2)
print(int(str3) + num3) # 17 : 정수값 -> str3 정수로 형변환 해줌!
print(str3 + str(num3)) # 107 : 문자열 -> num3 문자열로 형변환 해줌!
type(num3)
```

```python
num1 = 23
num2 = 3

print(f'더하기 결과 : {num1 + num2}')
print(f'빼기 결과 : {num1 - num2}')
print(f'곱하기 결과 : {num1 * num2}')
print(f'나누기 결과 : {(num1 / num2):.3f}')
print("나누기 결과 : {:.3f}".format(num1 / num2))
print("나누기 결과 : %.3f" % (num1 / num2))
```
