```python
num1 = 23
num2 = 3

print(f'  : {num1 + num2}')
print(f'  : {num1 - num2}')
print(f'  : {num1 * num2}')
print(f'  : {(num1 / num2):.3f}')
print("  : {:.3f}".format(num1 / num2))
print("  : %.3f" % (num1 / num2))
```
